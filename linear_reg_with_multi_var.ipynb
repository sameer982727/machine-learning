```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
sa=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv')
sa
```

```python
sa['test_score(out of 10)'].mean()
```

```python
sa['test_score(out of 10)']=sa['test_score(out of 10)'].fillna(sa['test_score(out of 10)'].mean())
sa
```

```python
sa.experience=sa.experience.fillna('zero')
sa
```

```python
!pip install word2number
from word2number import w2n
sa.experience=sa.experience.apply(w2n.word_to_num)
sa
```

```python
li=linear_model.LinearRegression()
li.fit(sa.drop(['salary($)'],axis='columns'),sa['salary($)'])
```
